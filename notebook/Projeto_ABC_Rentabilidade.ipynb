```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("SuperMarket Analysis.csv")
```

```python
df.columns = df.columns.str.lower().str.replace(" ", "_")
df = df.rename(columns={"tax_5%": "tax_5"})
```

```python
df_produto = df.groupby("product_line").agg({
    "sales": "sum",
    "cogs": "sum",
    "gross_income": "sum",
    "quantity": "sum"
}).reset_index()
```

```python
df_produto = df_produto.sort_values(
    by="gross_income",
    ascending=False
)
```

```python
df_produto["lucro_acumulado"] = df_produto["gross_income"].cumsum()

df_produto["percentual_acumulado"] = (
    df_produto["lucro_acumulado"] /
    df_produto["gross_income"].sum()
)
```

```python
def classificar_abc(p):
    if p <= 0.70:
        return "A"
    elif p <= 0.90:
        return "B"
    else:
        return "C"

df_produto["classe_abc"] = (
    df_produto["percentual_acumulado"]
    .apply(classificar_abc)
)
```

```python
df_produto["classe_abc"] = df_produto["classe_abc"].astype(str)
```

```python
df_produto.to_csv("dados_tratados.csv", index=False)
```
